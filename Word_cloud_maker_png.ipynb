```python
# Import libraries
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Baca file CSV
file_path = '/content/drive/MyDrive/sentimentbri/Analisis Sentimen BRI dan Livin- Hasil Analisis - BRI.csv'
df = pd.read_csv(file_path)

# Ambil konten dari kolom "Processed_Text"
processed_text = ' '.join(df['Processed_Text_3'].astype(str))

# Buat word cloud dengan penyesuaian
wordcloud = WordCloud(width=800, height=400, background_color='white', prefer_horizontal=1, colormap='viridis',
                      max_words=100, stopwords=None, contour_width=3, contour_color='steelblue',
                      scale=3, font_path='/content/drive/MyDrive/sentimentbri/DejaVu_Sans/DejaVuSans-Bold.ttf').generate(processed_text)

# Simpan word cloud
output_path = '/content/drive/MyDrive/sentimentbri/wordcloud/Word_Cloud_BRI.png'
wordcloud.to_file(output_path)

# Plot word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
